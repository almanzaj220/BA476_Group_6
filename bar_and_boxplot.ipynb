```python
import seaborn as sns

plt.figure(figsize = (20, 20))
cols = df.columns

for x in range(len(cols)):
  plt.subplot(6,2,x+1)
  sns.boxplot(df[cols[x]])
  plt.subplots_adjust(left=0.1,
                    bottom=0.1, 
                    right=0.5, 
                    top=0.9, 
                    wspace=0.4, 
                    hspace=0.4)

plt.show();
```
